```python
df_trans=df.copy()
```

```python
df_trans['t']= np.arange(1,43)
```

```python
df_trans['log']= np.log(df['Sales']) #log transform of data
```

```python
df_trans['t_sq']= df['Sales']*df['Sales'] # taking sq
```

```python
df_trans
```

```python
train= df_trans.head(int(len(df_trans['Sales'])*0.7))
train
```

```python
test= df_trans.tail(int(len(df_trans['Sales'])-len(train)))
```

```python
test
```

```python
#linear model
linear_model=smf.ols('Sales~t',data=train).fit()
pred_linear= pd.Series(linear_model.predict(pd.DataFrame(test['t'])))
rmse_linear=np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_linear))**2))
rmse_linear
```

```python
print('predicted data')
pred_linear.head(10)
```

```python
print('original data')
test['Sales'].head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
plt.title('Plot of predicted data vs original data')

# Plot original data in red
sns.lineplot(data=test, x=test.index, y='Sales', color='r', label='Original Data')

# Plot predicted data
sns.lineplot(data=pred_linear, label='Predicted Data')

plt.show()
```

```python
#quadratic
Quad = smf.ols('Sales~t+t_sq',data=train).fit()
pred_Quad = pd.Series(Quad.predict(test[['t','t_sq']]))
rmse_Quad=np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_Quad))**2))
rmse_Quad
```

```python
print('predicted data')
pred_Quad.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
plt.title('Plot of predicted data vs original data')

# Plot original data in red
sns.lineplot(data=test, x=test.index, y='Sales', color='r', label='Original Data')

# Plot predicted data
sns.lineplot(data=pred_Quad, label='Predicted Data (Quadratic)')

plt.show()
```

```python
#exponential
exp= smf.ols('log~t',data=train).fit()
pred_exp= pd.Series(exp.predict(pd.DataFrame(test['t'])))
rmse_exp=np.sqrt(np.mean((np.array(test['Sales'])-np.array(np.exp(pred_exp)))**2))
rmse_exp
```

```python
print('predicted data')
pred_exp.head(10)
```

```python
print('original data')
test['Sales'].head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
plt.title('Plot of predicted data vs original data')

# Plot original data in red
sns.lineplot(data=test, x=test.index, y='Sales', color='r', label='Original Data')

# Plot predicted data
sns.lineplot(data=pred_exp, label='Predicted Data (Exponential)')

plt.show()
```

```python
#log with Sq
exp= smf.ols('log~t+t_sq',data=train).fit()
pred_log_sq= pd.Series(exp.predict(test[['t','t_sq']]))
rmse_log_sq=np.sqrt(np.mean((np.array(test['log'])-np.array(pred_log_sq))**2))
rmse_log_sq
```

```python
print('predicted data')
pred_log_sq.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
plt.title('Plot of predicted data vs original data')

# Plot original log-transformed data in red
sns.lineplot(data=test, x=test.index, y='log', color='r', label='Original Data (log)')

# Plot predicted log-transformed data
sns.lineplot(data=pred_log_sq, label='Predicted Data (log-squared)')

plt.show()
```
