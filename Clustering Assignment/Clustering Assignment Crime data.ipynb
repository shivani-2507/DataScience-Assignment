# Kmeans

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
ds=pd.read_csv('crime_data.csv')
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
ds_norm=norm_func(ds.iloc[:,1:])
ds_norm
```

```python
# Elbow curv
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(ds_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(ds_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
ds['Clust']=x
```

```python
ds
```

```python
ds.iloc[:,1:5].groupby(ds.Clust).mean()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
da=pd.read_csv('crime_data.csv')
```

```python
da.info()
```

```python
dv=da.iloc[:,1:5]
```

```python
dv.values
```

```python
stscaler=StandardScaler().fit(dv.values)
x=stscaler.transform(dv.values)
```

```python
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
pd.concat([da,cl],axis=1)
```
